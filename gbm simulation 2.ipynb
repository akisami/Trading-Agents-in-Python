```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#drift coeffecient
mu = 0.1
# number of steps
n = 100
# time in years
T = 1
#number of sims
M=100
# initial stock price
S0 = 1
# volatility
sigma = 0.5


```

```python
# calc each time step
dt = T/n
#simulation using numpy arrayrs
St=np.exp(
    (mu-sigma**2/2)*dt
    + sigma*np.random.normal(0,np.sqrt(dt), size=(M,n)).T
)
#include arrays of 1's
St=np.vstack([np.ones(M),St])
#cumulative product
St=S0*St.cumprod(axis=0)

```

```python
#Define time interval 
time = np.linspace(0,T,n+1)
#Require numpy array that is the same as St
tt =np.full(shape=(M,n+1), fill_value=time).T
```

```python
plt.plot(tt, St)
plt.xlabel("Years $(t)$")
plt.ylabel("Stock Price $(S_t)$")
plt.title("GBM simulation")
plt.show()
```
